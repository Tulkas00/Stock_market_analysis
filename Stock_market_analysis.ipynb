```python
TESLA = yf.download('TSLA', start='2012-01-01', end='2017-01-01')
FORD = yf.download('F', start='2012-01-01', end='2017-01-01')
GM = yf.download('GM', start='2012-01-01', end='2017-01-01')
```

```python
TESLA.columns = [col[0] for col in TESLA.columns]
FORD.columns = [col[0] for col in FORD.columns]
GM.columns = [col[0] for col in GM.columns]
```

```python
TESLA.info()
```

```python
TESLA.Open.plot(label="Tesla")
GM.Open.plot(label="GM")
FORD.Open.plot(label="Ford")
plt.legend()
plt.ylabel("Stock Price")
```

```python
TESLA.Volume.plot(label="Tesla")
GM.Volume.plot(label="GM")
FORD.Volume.plot(label="Ford")
plt.legend()
plt.ylabel("Volume Traded")
```

####Market CAP

```python
TESLA['Total Traded']=TESLA['Open']*TESLA['Volume']
GM['Total Traded']=GM['Open']*GM['Volume']
FORD['Total Traded']=FORD['Open']*FORD['Volume']
```

```python
TESLA['Total Traded'].plot(label="Tesla",figsize=(15,7))
GM['Total Traded'].plot(label="GM")
FORD['Total Traded'].plot(label="FORD")
plt.legend()
plt.ylabel("Total Traded")
```

Moving average

```python
GM.Open.plot(label="GM",figsize=(15,7))
GM['MA50']=GM['Open'].rolling(50).mean()
GM.MA50.plot(label='MA50')
GM['MA200']=GM['Open'].rolling(200).mean()
GM.MA200.plot(label='MA200')
plt.legend()
```

Percentage change for volatility

```python
TESLA['returns']=(TESLA['Close']/TESLA['Close'].shift(1))-1
GM['returns']=(GM['Close']/GM['Close'].shift(1))-1
FORD['returns']=(FORD['Close']/FORD['Close'].shift(1))-1
```

```python
GM.head()
```

```python
TESLA['returns'].hist(bins=100,label='TESLA',figsize=(15))
GM['returns'].hist(bins=100,label='GM')
FORD['returns'].hist(bins=100,label='FORD')
plt.legend()
```

```python
TESLA['returns'].hist(bins=100,label='TESLA',figsize=(15,7))
GM['returns'].hist(bins=100,label='GM')
FORD['returns'].hist(bins=100,label='FORD')
plt.legend()
```

#Cumulative Return

```python
TESLA['Cumulative returns']=(1+TESLA['returns']).cumprod()
GM['Cumulative returns']=(1+ GM['returns']).cumprod()
FORD['Cumulative returns']=(1+ FORD['returns']).cumprod()
```

```python
TESLA['Cumulative returns'].plot(label='TESLA',figsize=(15,7))
GM['Cumulative returns'].plot(label='GM')
FORD['Cumulative returns'].plot(label='FORD')
plt.legend()
```
